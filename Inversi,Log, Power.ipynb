```python
#import library
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
#membaca gambar dari directory
img = imageio.imread("fototegar2.jpeg")
```

```python
#Menambahkan resolusi tinggi dan lebar sebuah gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
```

Inversi

```python
img_inversi = np.zeros(img.shape, dtype=np.uint8) #membuat array kosong
```

```python
def inversi_grayscale(nilai): #mengubah nilai grayscale (skala abu-abu) sebuah piksel menjadi nilai kebalikannya.
    #melakukan iterasi untuk setiap nilai pada sumbul x,y
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0] #Mengambil nilai intensitas warna merah pada pixel 
            green = img[y][x][1] #Mengambil nilai intensitas warna hijau pada pixel 
            blue = img[y][x][2] #Mengambil nilai intensitas warna biru pada pixel 
            gray = (int(red) + int(green) + int(blue)) / 3  #berfungsi untuk menghitung rata-rata dari intensitas ketiga komponen warna (merah, hijau, dan biru) dari suatu piksel dalam gambar berwarna
            gray = nilai - gray #mengurangi nilai dengan rata rata intensitas dari abu-abu
            img_inversi[y][x] = (gray, gray, gray) #berfungsi untuk mengubah piksel pada koordinat (x,y) dalam gambar img_inversi dengan nilai intensitas keabuan gray.
```

```python
def inversi_rgb(nilai): #mengubah nilai grayscale (skala abu-abu) sebuah piksel menjadi nilai kebalikannya.
    #melakukan iterasi untuk setiap nilai pada sumbul x,y
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0] #Mengambil nilai intensitas warna merah pada pixel 
            red = nilai - red  #mengurangi nilai dengan rata rata intensitas dari merah
            green = img[y][x][1] #Mengambil nilai intensitas warna hijau pada pixel 
            green = nilai - green  #mengurangi nilai dengan rata rata intensitas dari hijau
            blue = img[y][x][2] #Mengambil nilai intensitas warna merah pada pixel 
            blue = nilai - blue  #mengurangi nilai dengan rata rata intensitas dari biru
            img_inversi[y][x] = (red, green, blue) #berfungsi untuk mengubah piksel pada koordinat (x,y) dalam gambar img_inversi dengan nilai intensitas RGB.
```

```python
inversi_grayscale(255) #inisialisasi array
plt.imshow(img_inversi) #menampilkan gambar yang sudah disimpan
plt.title("Inversi Grayscale") #menambahkan judul
plt.show() #library matplotlib untuk menampilkan gambar

inversi_rgb(255) #inisialisasi array
plt.imshow(img_inversi) #menampilkan gambar yang sudah disimpan
plt.title("Inversi RGB") #menambahkan judul
plt.show() #library matplotlib untuk menampilkan gambar
```

Log

```python
img_log = np.zeros(img.shape, dtype=np.uint8) #membuat array kosong
```

```python
def log(c): #menambahkan fungsi dengan mengambil parameter c
    for y in range(0, img_height): #melakukan iterasi dari 0 hingga tinggi gambar
        for x in range(0, img_width): #melakukan iterasi dari 0 hingga lebar gambar
            red = img[y][x][0] #Mengambil nilai intensitas warna merah pada pixel 
            green = img[y][x][1] #Mengambil nilai intensitas hijau pada pixel 
            blue = img[y][x][2] #Mengambil nilai intensitas warna biru pada pixel 
            gray = (int(red) + int(green) + int(blue)) / 3 #berfungsi untuk menghitung rata-rata dari intensitas ketiga komponen warna (merah, hijau, dan biru) dari suatu piksel dalam gambar berwarna
            gray = int(c * np.log(gray + 1)) #Mengubah nilai intensitas keabuan dengan mengambil logaritma natural dari nilai intensitas keabuan yang telah dihitung pada baris sebelumnya
            if gray > 255: #Memastikan nilai intensitas keabuan yang baru tidak melebihi 255
                gray = 255
            if gray < 0: #Memastikan nilai intensitas keabuan yang baru tidak kurang dari 0
                gray = 0
            img_log[y][x] = (gray, gray, gray) #Menyimpan nilai intensitas keabuan yang baru ke dalam citra hasil dengan cara mengubah nilai intensitas merah, hijau, dan biru menjadi nilai intensitas keabuan yang sama
```

```python
log(30) #mendefinisikan nilai c pada log adalah 30
plt.imshow(img_log) #menampilkan gambar yang sudah disimpan
plt.title("Log") #membuat judul
plt.show() #library matplotlib untuk menampilkan gambar
```

Nth Power

```python
img_nthpower = np.zeros(img.shape, dtype=np.uint8)
```

```python
def nthpower(c, y): #Membuat fungsi dengan nama "nthpower" dengan 2 parameter masukan yaitu "c" dan "y"
    #Membuat variabel "thc" dan "thy" dengan nilai c/100 dan y/100
    thc = c / 100
    thy = y / 100
    #Melakukan iterasi untuk setiap baris dan kolom pada gambar
    for y in range(0, img_height):
        for x in range(0, img_width):
            #Memperoleh nilai warna merah, hijau, dan biru (RGB) pada setiap piksel pada posisi (y, x) pada gambar.
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 #Menghitung nilai keabuan pada setiap piksel dengan cara menjumlahkan nilai RGB, kemudian dibagi 3.
            gray = int(thc * pow(gray, thy)) #Menghitung nilai keabuan setiap piksel dengan memperhatikan faktor pengali c dan y yang diberikan pada parameter masukan
            #Memastikan bahwa nilai keabuan setiap piksel berada dalam rentang nilai yang valid (yaitu antara 0 dan 255)
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray) #Menyimpan nilai keabuan yang telah dihitung pada setiap piksel pada posisi (y, x)
```

```python
nthpower(50, 100) #mendefinisikan nilai c dan y 
plt.imshow(img_nthpower) #menampilkan gambar yang sudah disimpan
plt.title("Nth Power") #menambahkan judul
plt.show() #library matplotlib untuk menampilkan gambar
```

Nth Root Power

```python
img_nthrootpower = np.zeros(img.shape, dtype=np.uint8) #membuat array kosong
```

```python
#Mendefinisikan fungsi dengan nama "nthrootpower" yang menerima dua argumen yaitu "c" dan "y".
def nthrootpower(c, y):
    #Menghitung nilai "thc" dan "thy" sebagai hasil dari "c" dan "y" dibagi dengan 100.
    thc = c / 100
    thy = y / 100
    #Mengulang perintah untuk setiap koordinat "y" dan "x" pada gambar.
    for y in range(0, img_height):
        for x in range(0, img_width):
            #Mengambil nilai pixel warna merah, hijau, dan biru (RGB) pada koordinat "y" dan "x" dari gambar.
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            #Menghitung nilai keabuan (grayscale) sebagai rata-rata dari tiga nilai warna RGB di atas.
            gray = (int(red) + int(green) + int(blue)) / 3
            #Menghitung nilai keabuan yang telah diubah dengan perpangkatan n (n sesuai dengan nilai "y" yang dimasukkan saat memanggil fungsi "nthrootpower"). 
            gray = int(thc * pow(gray, 1./thy))
            #Menyaring nilai keabuan agar tidak melebihi 255 atau kurang dari 0.
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray) #Mengganti nilai pixel gambar dengan nilai keabuan yang telah diubah.
```

```python
nthrootpower(50, 100) #Memanggil fungsi "nthrootpower" dengan nilai "c" = 50 dan "y" = 100.
#Menampilkan gambar dengan nilai keabuan yang telah diubah menggunakan "nthrootpower". Judul gambar tersebut adalah "Nth Root Power".
plt.imshow(img_nthrootpower) 
plt.title("Nth Root Power")
plt.show()
```
