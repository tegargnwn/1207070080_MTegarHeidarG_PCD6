```python
#import library
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("fototegar2.jpeg") #Membaca file gambar dan mengubahnya menjadi representasi numerik
```

```python
#Menambahkan resolusi tinggi dan lebar sebuah gambar dan type gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
```

```python
img_grayscale = np.zeros(img.shape, dtype=np.uint8) #membuat array kosong

for y in range(0, img_height): #looping untuk setiap piksel pada dimensi vertikal 
    for x in range(0, img_width): #looping untuk setiap piksel pada dimensi horizontal
        red = img[y][x][0] #mengambil nilai intensitas merah pada koordinat (y,x)
        green = img[y][x][1] #mengambil nilai intensitas hijau pada koordinat (y,x)
        blue = img[y][x][2] #mengambil nilai intensitas biru pada koordinat (y,x)
        gray = (int(red) + int(green) + int(blue)) / 3 #berfungsi untuk menghitung rata-rata dari intensitas ketiga komponen warna (merah, hijau, dan biru) dari suatu piksel dalam gambar berwarna
        img_grayscale[y][x] = (gray, gray, gray) #berfungsi untuk mengubah piksel pada koordinat (x,y) dalam gambar img_grayscale dengan nilai intensitas keabuan gray.
        
plt.imshow(img_grayscale) #menampilkan gambar
plt.title("Grayscale") #menambahkan judul
plt.show() #library matplotlib untuk menampilkan gambar
```

Menampilkan Histogram Gambar Grayscale

```python
hg = np.zeros((256)) #Membuat array kosong
```

```python
for x in range(0, 256): #berfungsi untuk melakukan loop atau perulangan sebanyak 256 kali
    hg[x] = 0 #berfungsi untuk mengatur nilai elemen ke-x dalam array hg menjadi 0.
```

```python
for y in range(0, img_height): #looping untuk setiap piksel pada dimensi vertikal 
    for x in range(0, img_width): #looping untuk setiap piksel pada dimensi horizontal
        gray = img_grayscale[y][x][0] #mengambil nilai intensitas keabuan pada piksel dengan koordinat (x,y)
        hg[gray] += 1 #mengatur nilai elemen x gray
```

```python
# plt.figure(figsize=(20, 6))
# plt.plot(hg, color="black", linewidth=2.0)
# plt.show()

bins = np.linspace(0, 256, 100) #membuat array satu dimensi yang terdiri dari 100 bilangan yang terdistribusi merata antara 0 dan 256.
plt.hist(hg, bins, color="black", alpha=0.5) #membuat histogram intensitas keabuan gambar.
plt.title("Histogram") #menambahkan judul
plt.show() #library matplotlib untuk menampilkan gambar
```

Menampilkan Histogram Kumulatif

```python
#inisialisasi array
hgk = np.zeros((256)) 
c = np.zeros((256))

#melakukan iterasi
for x in range(0, 256):
    hgk[x] = 0
    c[x] = 0

#melakukan iterasi untuk setiap nilai pada sumbul x,y
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0] #mengambil nilai grayscale dari RGB
        hgk[gray] += 1 #menambahkan nilai array dengan 1
                
c[0] = hgk[0] #menambakan nilai pertama pada array pada hgk
for x in range(1, 256): #melakukan iterasi 
     c[x] = c[x-1] + hgk[x] #menambahkan nilai array

hmaxk = c[255] #menambahkan nilai elemen terakhir

for x in range(0, 256): #melakukan iterasi pada sumbu x
    c[x] = 190 * c[x] / hmaxk #nilai array dikalikan

plt.hist(c, bins, color="black", alpha=0.5) #menampilkan histogram
plt.title("Histogram Grayscale Kumulatif") #memberikan judul
plt.show() #menampilkan histogram pada jendela grafik
```

Menampilkan Histogram Hequalisasi

```python
#inisialisasi array
hgh = np.zeros((256))
h = np.zeros((256))
c = np.zeros((256))

#melakukan iterasi
for x in range(0, 256):
    hgh[x] = 0
    h[x] = 0
    c[x] = 0

#melakukan iterasi untuk setiap nilai pada sumbul x,y
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hgh[gray] += 1

#menambahkan nilai array                
h[0] = hgh[0]
for x in range(1, 256):
     h[x] = h[x-1] + hgh[x]

#melakukan iterasi
for x in range(0, 256):
     h[x] = h[x] / img_height / img_width

#melakukan iterasi
for x in range(0, 256):
    hgh[x] = 0

#melakukan iterasi untuk setiap nilai pada sumbul x,y    
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        gray = h[gray] * 255
        hgh[int(gray)] += 1

#menambahkan nilai array
c[0] = hgh[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgh[x]

hmaxk = c[255] #mengambil nilai dari elemen ke-255 dari array c

for x in range(0, 256): #Loop for ini digunakan untuk melakukan iterasi pada setiap intensitas piksel grayscale mulai dari 0 hingga 255. 
    c[x] = 190 * c[x] / hmaxk
plt.hist(c, bins, color="black", alpha=0.5) #menampilkan histogram
plt.title("Histogram Grayscale Hequalisasi") #Memberikan judul pada histogram equalization
plt.show() #Menampilkan histogram equalization pada layar
```
