```python
#import library
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
#membaca gambar
img = imageio.imread("fototegar.jpeg")
```

```python
img_height = img.shape[0] #mengambil ketinggian gambar dari dimensi array
img_width = img.shape[1] #mengambil lebar gambar dari dimensi array
img_channel = img.shape[2] #mengambil jumlah channel gambar dari dimensi array
img_type = img.dtype #mengambil tipe data gambar dari dimensi array
```

```python
img_flip_horizontal = np.zeros(img.shape, img_type) #membuat array kosong dengan posisi horizontal
img_flip_vertical = np.zeros(img.shape, img_type)  #membuat array kosong dengan posisi vertikal
```

```python
for y in range(0, img_height): #looping untuk setiap piksel pada dimensi vertikal 
    for x in range(0, img_width): #looping untuk setiap piksel pada dimensi horizontal
        for c in range(0, img_channel): #looping untuk setiap channel warna
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] #Melakukan flipping horizontal pada gambar
```

```python
for y in range(0, img_height): #looping untuk setiap piksel pada dimensi vertikal 
    for x in range(0, img_width): #looping untuk setiap piksel pada dimensi horizontal
        for c in range(0, img_channel): #looping untuk setiap channel warna
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] #Melakukan flipping vertikal pada gambar
```

```python
plt.imshow(img_flip_horizontal) #menampilkan gambar dengan posisi horizontal
plt.title("Flip Horizontal") #menambahkan judul
plt.show() #library matplotlib untuk menampilkan gambar
plt.imshow(img_flip_vertical) #menampilkan gambar dengan posisi horizontal
plt.title("Flip Vertical") #menambahkan judul
plt.show() #library matplotlib untuk menampilkan gambar
```
